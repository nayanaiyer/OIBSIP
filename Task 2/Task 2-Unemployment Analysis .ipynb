In both the datasets, we convert the 'Date' column to datetime and 'Frequency' and 'Region' columns to categorical columns

```python
data1['Date']=pd.to_datetime(data1['Date'],dayfirst=True)
data1['Frequency'] = data1['Frequency'].astype('category')
data1['Region'] =data1['Region'].astype('category')
data1.info()
```

We extract the Month from the Date in a new column 'Month' in both the datasets

```python
#Create new month column 
data1['Month']=data1['Date'].dt.month
#Convert Month to Month Name
data1['Month Name']=data1['Month'].apply(lambda x: calendar.month_abbr[x])
data1
```

```python
#Create new month column 
data2['Month']=data2['Date'].dt.month
#Convert Month to Month Name
data2['Month Name']=data2['Month'].apply(lambda x: calendar.month_abbr[x])
data2
```

We will split the dataset data1 into two parts: One before lockdown(Date Range: 2019-05-31' to '2020-03-31'and second one 2020-04-01 onwards.

```python
# We create a boolean variable 'mask' which will return true if it matches the date range 
mask1 = (data1['Date'] >= '2019-05-31') & (data1['Date'] <= '2020-03-31')
data1_new = data1.loc[mask1]
data1_new
```

```python
# We create a boolean variable 'mask' which will return true if it matches the date range 
mask2 = (data2['Date']>= '2020-04-30')
data2_new = data2.loc[mask2]
data2_new
```

# 3. Data Visualization

Visualizing the correlation of Stats before Lockdown and after lockdown

```python
mpl.figure(figsize=(30,10))
mpl.subplot(1,2,1)
mpl.title("Correlation BEFORE Lockdown")
sb.heatmap(data=data1_stats.corr(),annot=True,cmap='coolwarm')

mpl.subplot(1,2,2)
mpl.title("Correlation During Lockdown")
sb.heatmap(data=data2_stats.corr(),annot=True,cmap='coolwarm')
mpl.show()
```

Plotting Graphs to Visualize distribution between Rural and Urban areas (Dataset 1) and Different regions(dataset 2)

```python
mpl.figure(figsize=(25,10))

#Graph 1
mpl.subplot(1,2,1)

sb.barplot(data=data1_new,x='Region',y='Estimated Unemployment Rate (%)',hue='Region',ci=None)

mpl.xlabel("Area",fontsize=16)
mpl.ylabel("Estimated Unemployment Rate (%)", fontsize=16)
mpl.title("Unemployment before lockdown(Area wise)",fontsize=16)

#Graph 2
mpl.subplot(1,2,2)

sb.barplot(data=data2_new, x='Region', y='Estimated Unemployment Rate (%)',ci=None,palette='Paired_r')

mpl.xticks(rotation=90)
mpl.xlabel("Region",fontsize=16)
mpl.ylabel("Estimated Unemployment Rate (%)", fontsize=16)
mpl.title("Unemployment after lockdown(Region wise)",fontsize=16)
```

Plotting State-wise Bar graph to find Unemployment Rate before and During lockdown

```python
#Statewise Stats
state_stats1=round(data1_new.groupby(['States'])
                   [['Estimated Unemployment Rate (%)','Estimated Employed',
                                         'Estimated Labour Participation Rate (%)']].mean().reset_index(),2)
state_stats1=state_stats1.sort_values(by='Estimated Unemployment Rate (%)',ascending=False)
```

```python
state_stats1.head()
```

```python
state_stats1.tail()
```

```python
state_stats2=round(data2_new.groupby(['States'])
                   [['Estimated Unemployment Rate (%)','Estimated Employed',
                                         'Estimated Labour Participation Rate (%)']].mean().reset_index(),2)
state_stats2=state_stats2.sort_values(by='Estimated Unemployment Rate (%)',ascending=False)
```

```python
state_stats2.head()
```

```python
state_stats2.tail()
```

```python
mpl.figure(figsize=(16,9))

#Before Lockdown
mpl.subplot(1,2,1)
mpl.title("State wise Unemployment Rate before lockdown")
mpl.xlabel("States")
mpl.ylabel("Estimated Unemployment Rate (%)")
mpl.xticks(rotation=90)

sb.barplot(data=state_stats1,x='States',y='Estimated Unemployment Rate (%)',ci=None)


#During lockdown
mpl.subplot(1,2,2)
mpl.title("State wise Unemployment Rate During lockdown")
mpl.xlabel("States")
mpl.ylabel("Estimated Unemployment Rate (%)")
mpl.xticks(rotation=90)

sb.barplot(data=state_stats2,x='States',y='Estimated Unemployment Rate (%)',ci=None)
```
